# Demo: defining a class

```
class taxicab:
# A class representing a taxicab
  def __init__(self , n, a):
    self.year = n
    self.make = a
  def get_make(self):
    return self.make
```

# Demo: creating class instances

```
cab_1 = taxicab(2015, "Chevrolet")
cab_2 = taxicab(2017, "Ford")

print(cab_1.get_make())
print(cab_2.get_make())
```

# Demo: class inheritance

```
class rental_car(taxicab):
    def __init__(self,n,a):
        taxicab.__init__(self,n,a)

rental_1 = rental_car(2016, "Dodge")
print(rental_1.get_make())
```
